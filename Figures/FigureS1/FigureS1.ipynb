```python
import glob
import pandas as pd
import numpy as np
```

#### Plotting for T9SS proteins 

```python
df1 = pd.read_csv('Protein_component_count_T9SS_gliding_motility.csv',sep='\t')
print(df1.shape)
df1.head()
```

```python
T9SS_count = list(df1['T9SS_Proteins'])
```

```python
# Histogram for T9SS proteins
import matplotlib.pyplot as plt

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams.update({'font.family':'sans-serif'})
plt.rc('legend',fontsize=6)
plt.rc('axes',labelsize=5)
plt.rc('xtick', labelsize=8)
plt.rc('ytick', labelsize=8)

fig, ax = plt.subplots(figsize=(3.3,2), dpi=600)
bins = np.arange(10,24)-0.5
ax.hist(T9SS_count, bins,facecolor = '#fce1e4', edgecolor = 'grey')


ax.set_xlabel('Number of Proteins')

ax.set_ylabel('Number of Bacteroidetes')


ax.set_xlim(xmin=9, xmax=23)
start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start+1, end+1, 2))

ax.set_ylim(ymin=0, ymax=140)
start, end = ax.get_ylim()
ax.yaxis.set_ticks(np.arange(start, end+1, 20))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.tight_layout()

fig.savefig('T9SS_histogram.pdf')
```

#### Plotting for gliding motility

```python
GM_count = df1['Gliding_Motility_proteins']
```

```python
# histogram plot for gliding motility
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams.update({'font.family':'sans-serif'})
plt.rc('legend',fontsize=6)
plt.rc('axes',labelsize=5)
plt.rc('xtick', labelsize=8)
plt.rc('ytick', labelsize=8)

fig, ax = plt.subplots(figsize=(3.3,2), dpi=600)
bins = np.arange(18,27)-0.5
ax.hist(GM_count, bins,facecolor ='#e8dff5', edgecolor = 'gray')


ax.set_xlabel('Number of Proteins')

ax.set_ylabel('Number of Bacteroidetes')


ax.set_xlim(xmin=17, xmax=26)
start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start+1, end+1, 1))

ax.set_ylim(ymin=0, ymax=70)
start, end = ax.get_ylim()
ax.yaxis.set_ticks(np.arange(start, end+1, 10))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.tight_layout()

fig.savefig('GM_histogram.pdf')
```
